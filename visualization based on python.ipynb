```python

colnames=['Magnitude', 'Time', 'Latitude', 'Longitude', 'Depth', 'Location']
df = pd.read_excel('earthquake.xlsx',skiprows=0,names=colnames)
X = df.Latitude
Y = df.Magnitude
Z = df.Depth
def A():
    plt.scatter(X,Y,color="blue",linewidth=2)
    plt.title("RM scatter",color="blue")
    plt.grid()
    plt.show()
A()
```

```python
from scipy.optimize import leastsq
def func(p,x):
    a,b,c=p
    return a*x*x+b*x+c
def error(p,x,y):
    return func(p,x)-y
def main():
    plt.figure(figsize=(10,6))
    p0=[0,0,0]
    Para = leastsq(error,p0,args=(X,Y))
    a,b,c=Para[0]
    print("a=",a,"b=",b,"c=",c)
    plt.scatter(X,Y,color="blue",linewidth=2)
    x=np.linspace(0,20,20)
    y=a*x*x+b*x+c
    plt.plot(x,y,color="blue",linewidth=2,)
    plt.title("地震深度分布")
    plt.grid()
    plt.show()

main()
```
